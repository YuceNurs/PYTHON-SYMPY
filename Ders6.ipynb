-Eğer belirtiyorsa "verilen kenarlar bir üçgen belirtir ve bu üçgenin çevresi = ?? birimdir. Eğer belirtmiyorsa der ki bro whats your problem.. "

```python
c= int(input("Bir uzunluk giriniz"))
b= int(input("Bir uzunluk giriniz"))
a= int(input("Bir uzunluk giriniz"))

def ucgen(a,b,c):
     if a+b <= c or a+c <=b or b+c <=a: 
        return ("Bro whats your problem..")
     else :
         return ("Bu bir üçgen belirtir ve bu üçgenin çevresi: " , a+b+c)
```
