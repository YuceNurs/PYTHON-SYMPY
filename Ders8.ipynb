```python
from sympy import*
from IPython.display import display
from sympy.plotting import plot, plot3d, plot3d_parametric_line, plot3d_parametric_surface
from sympy import init_printing
init_printing(use_latex="mathjax")
```

```python
x,y,z = symbols ("x y z")
```

```python
f1= x**2
display(f1)
plot(f1)
```

```python
f2= x**5+x**3+17
display(f2)
plot(f2)
```

```python
plot(f1, (x,-5,5))
```

```python
p= plot(f1, show= False)
p.xlim = (-5,5)
p.ylim = (-5,5)
p.show()
```

```python
p= plot(f1, show= False)
p.xlim = (-5,5)
p.ylim = (-5,5)
p.title ="Parabole Graph"
p.xlabel= "x label"
p.ylabel = "y label"
p.show()
```

```python
plot(x, x**2, x**3, (x,-3,3))
```

```python
p2 = plot(x, x**2,x**3, show =False)
p2.xlim= (-2,3)
p2.ylim=(-3,3)
p2[0].line_color= "red"
p2[1].line_color= "blue"
p2[2].line_color= "green"
p2.show()
```

```python
f3= Piecewise((x**2,x<1),(x+1,x>1))
f3
```

```python
plot(f3,(x,-3,3))
```

```python
f4= Piecewise((x**2,x<1),(x+1,True))
f4
```

```python
plot(f4,(x,-3,3))
```

```python
f5=x**2+y**2
```

```python
plot3d(f5,(x,-5,5), (y,-5,5))
```

```python
plot3d(x**2+y**2, x**2-y**2,(x,-5,5), (y,-5,5))
```

```python
plot3d(x**2-y**2)
```

```python
t= symbols("t")
```

```python
plot3d_parametric_line(t,t**2,sin(t),(t,-4,4))
```

```python
u,v =symbols("u v ")
plot3d_parametric_surface(u,v,cos(u), (u,-4,4), (v,-4,4))
```
