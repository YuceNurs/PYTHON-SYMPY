## *args Kullanımı 

```python
def topla3(*sayilar):
    return sum(sayilar)
```

```python
def islemler (a,b):
    toplam = a+b
    fark = a-b
    carpim = a*b
    bolum = a/b
    return toplam , fark, carpim, bolum
```

```python
islemler(215412,6632324)
```

```python
def islemler2 (a,b):
    toplam = a+b
    fark = a-b
    carpim = a*b
    bolum = a/b
    print(f"Toplam = {toplam}")
    print(f"Fark = {fark}")
    print(f"Carpim = {carpim}")
    print(f"Bolum = {bolum}")
```

```python
islemler2(425435212451,5341234532)
```

```python
def islem3 (a):
    kare = a**2
    kup = a**3
    print(f"Karesi = {kare}")
    print(f"Kup = {kup}")
```

```python
islem3(23323221523215421251014.25321)
```

```python
def faktoriyel(n):
    if n <0:
        return ("hesap hatası sayıyı pozitif seçiniz")
    sonuc = 1 
    for i in range (1,n+1):
        sonuc = sonuc*i
    print(f"{n}!= {sonuc}")
```

```python
faktoriyel(-5)
```

```python
faktoriyel(105)
```
