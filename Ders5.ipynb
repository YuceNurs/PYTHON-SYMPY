```python
from sympy import *
from IPython.display import display
from sympy import init_printing
init_printing(use_latex='mathjax')

# Sembolik değişken tanımı
x, y = symbols('x y')

# 1) Polinom ve türev
f = x**3 + 2*x**2 - x + 5
display(f)
display(diff(f, x))

# 2) İntegral
g = sin(x) * exp(x)
display(integrate(g, x))

# 3) Limit
h = (sin(x)) / x
display(limit(h, x, 0))

# 4) Lineer denklem çözümü
eq1 = Eq(2*x + y, 5)
eq2 = Eq(x - y, 1)
sol = solve((eq1, eq2), (x, y))
display(sol)

# 5) Matris işlemleri
A = Matrix([[1, 2], [3, 4]])
display(A)
display(A.det())
display(A.inv())
```
