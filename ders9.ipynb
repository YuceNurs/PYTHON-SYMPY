```python
def limit_al(fonksiyon,degisken,nokta):
    lim=Limit(fonksiyon,degisken,nokta, "+-")
    cevap=lim.doit()
    sonuc= Eq(lim,cevap)
    display(sonuc)
```

```python
limit_al(fonk1, x, -1)
```

```python
fonk3=(2*x**2+1)/(x**2+3)
fonk3
```

```python
fonk4= 1/x
fonk4
limit_al(fonk4,x,0)
```

```python
def limit_al2(fonksiyon,degisken,nokta,sag_sol):
    lim=Limit(fonksiyon,degisken,nokta, sag_sol)
    cevap=lim.doit()
    sonuc= Eq(lim,cevap)
    display(sonuc)
```

```python
limit_al2(fonk4,x,0,"+")
```
